### Generate plot using Artist Layer

```python
#Import libraries
from matplotlib.backends.backend_agg import FigureCanvasAgg as FigureCanvas
from matplotlib.figure import Figure
import numpy as np

# Initializat figure and canvas
fig = Figure()
canvas = FigureCanvas(fig)

# Create 10000 random numbers using numpy
x = np.random.randn(10000)

# Create an axis artist
ax = fig.add_subplot(111)

# Generate histogram of 10000 numbers
ax.hist(x, 100)

# Add a title to the figure and save it
ax.set_title('Normal distribution with $\mu=0, \sigma=1$')

# Save plot as PNG
fig.savefig('matplotlib_histogram1.png')
```

### Generate plot using Scripting Layer

```python
#Import libraries
import matplotlib.pyplot as plt
import numpy as np

# Create 10000 random numbers using numpy
x = np.random.randn(10000)

# Generate histogram of 10000 numbers
plt.hist(x, 100)

# Add a title to the figure and save it
plt.title(r'Normal distribution with $\mu=0, \sigma=1$')

# Save plot as PNG
plt.savefig('matplotlib_histogram2.png')
plt.show()
```

### Generate plot using Pandas

```python
#Import libraries
import numpy as np
import pandas as pd

# Create a dataframe
df = pd.DataFrame(np.random.randn(10))

# plot dataframe as line graph
df.plot(kind='line')

# plot dataframe as histogram
df.plot(kind='hist')
```
